### Connect to the database

Let us now load the ipython-sql  extension and establish a connection with the database

##### The syntax for connecting to magic sql using sqllite is

  **%sql sqlite://DatabaseName**   
  
where DatabaseName will be your **.db** file 


```python
import sqlite3
import pandas as pd
con = sqlite3.connect("RealWorldData.db")
cur = con.cursor()
```

```python
%load_ext sql
```

```python
%sql sqlite:///RealWorldData.db
```

### Store the dataset in a Table

##### In many cases the dataset to be analyzed is available as a .CSV (comma separated values) file, perhaps on the internet. To analyze the data 
##### using SQL, it first needs to be stored in the database.

##### We will first read the csv files  from the given url  into pandas dataframes

##### Next we will be using the  df.to_sql() function to convert each csv file  to a table in sqlite  with the csv data loaded in it.


```python
df = pd.read_csv('./Chicago_Public_Schools_-_Progress_Report_Cards__2011-2012_.csv')
df.head()
```

```python
df.to_sql('RealWorldData', con, if_exists='replace', index=False)
```

### Problem 9

##### Get the 5 Community Areas with the least total College Enrollment  sorted in ascending order


```python
college_enroll_by_community_area = df.groupby('Community Area Name').agg(
    total_enrollment=('College Enrollment (number of students) ',
                      'sum')).reset_index()
college_enroll_by_community_area.nsmallest(
    5, 'total_enrollment')[[
        'Community Area Name', 'total_enrollment'
    ]].sort_values('total_enrollment')
```

```python
%sql select [Community Area Name], sum([College Enrollment (number of students) ]) AS TOTAL_ENROLLMENT \
   from realworlddata \
   group by [Community Area Name] \
   order by TOTAL_ENROLLMENT asc \
   LIMIT 5
```

Double-click **here** for a hint

### Problem 11

##### Get the hardship index for the community area which has College Enrollment of 4368


```python
social = pd.read_csv('https://data.cityofchicago.org/resource/jcxq-k9xf.csv')
social.to_sql("chicago_socioeconomic_data", con, if_exists='replace', index=False,method="multi")
social_college = df.merge(social,  left_on='Community Area Number', right_on='ca')
social_college[social_college['College Enrollment (number of students) '] == 4368][['Community Area Name', 'hardship_index']]
```

```python
%%sql 
select hardship_index from chicago_socioeconomic_data CD, realworlddata CPS 
where CD.ca = [Community Area Number]
and [College Enrollment (number of students) ] = 4368
```

Double-click **here** for the solution.

<!-- Solution:

### Problem 12

##### Get the hardship index for the community area which has the highest value for College Enrollment


```python
social_college.nlargest(1, 'College Enrollment (number of students) ')[['community_area_name', 'hardship_index']]
```

```python
%sql select ca, community_area_name, hardship_index from chicago_socioeconomic_data \
   where ca in \
   ( select [community area number] from realworlddata order by [College Enrollment (number of students) ] desc limit 1 )
```
